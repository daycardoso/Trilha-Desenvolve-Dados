### Algumas instalações necessarias

```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
```

### Importando o DataSet que vamos usar

```python
df = pd.read_csv('https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv')
df
```

### Um pouco de pré-processamento de dados

```python
# Preencher NAs com 0s e criar coluna PIB per capita
df = df.fillna(0)
df['pbi_per_capita'] = np.where(df['population']!= 0, df['gdp']/ df['population'], 0)
```

```python
# Tornando o pipeline de DataFrame interativo
idf = df.interactive()
```

### Emissão de CO2 ao longo do tempo por continente 

```python
# Definir widgets do painel
year_slider = pn.widgets.IntSlider(name='Year slider', start=1750, end=2020, step=5, value=2015)
```

```python
year_slider
```

```python
# Botões de rádio para medidas de CO2
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['co2', 'co2_per_capita',],
    button_type='success'
)
```

```python
continents = ['World', 'Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_pipeline = (
    idf[
        (idf.year <= year_slider) &
        (idf.country.isin(continents))
    ]
    .groupby(['country', 'year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_pipeline
```

```python
co2_plot = co2_pipeline.hvplot(x = 'year', by='country', y=yaxis_co2,line_width=2, title="CO2 emission by continent")
co2_plot
```

### Tabela - Emissão de CO2 ao longo do tempo por continente

```python
co2_table = co2_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width') 
co2_table
```

### Gráfico de dispersão de CO2 vs PIB

```python
co2_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.year == year_slider) &
        (~ (idf.country.isin(continents)))
    ]
    .groupby(['country', 'year', 'pbi_per_capita'])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_vs_gdp_scatterplot_pipeline
```

```python
co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(x='pbi_per_capita', 
                                                                y='co2', 
                                                                by='country', 
                                                                size=80, kind="scatter", 
                                                                alpha=0.7,
                                                                legend=False, 
                                                                height=500, 
                                                                width=500)
co2_vs_gdp_scatterplot
```

### Gráfico de barras com fontes de CO2 por continente

```python
yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['coal_co2', 'oil_co2', 'gas_co2'], 
    button_type='success'
)

continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_source_bar_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.country.isin(continents_excl_world))
    ]
    .groupby(['year', 'country'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(kind='bar', 
                                                     x='country', 
                                                     y=yaxis_co2_source, 
                                                     title='Fonte de CO2 por continente')
co2_source_bar_plot
```

### Criando Dashboard

```python
#Layout using Template
template = pn.template.FastListTemplate(
    title='Dashboard mundial de emissões de CO2', 
    sidebar=[pn.pane.Markdown("# Emissões de CO2 e Mudanças Climáticas"), 
             pn.pane.Markdown("#### As emissões de dióxido de carbono são o principal fator da mudança climática global. É amplamente reconhecido que, para evitar os piores impactos das mudanças climáticas, o mundo precisa reduzir urgentemente as emissões. Mas como essa responsabilidade é compartilhada entre regiões, países e indivíduos tem sido um ponto de discórdia interminável nas discussões internacionais."), 
             pn.pane.PNG('climate_day.png', sizing_mode='scale_both'),
             pn.pane.Markdown("## Configurações"),   
             year_slider],
    main=[pn.Row(pn.Column(yaxis_co2, 
                           co2_plot.panel(width=700), margin=(0,25)), 
                 co2_table.panel(width=500)), 
          pn.Row(pn.Column(co2_vs_gdp_scatterplot.panel(width=600), margin=(0,25)), 
                 pn.Column(yaxis_co2_source, co2_source_bar_plot.panel(width=600)))],
    accent_base_color="#F08080",
    header_background="#F08080",
)
template.show()
template.servable();
```
